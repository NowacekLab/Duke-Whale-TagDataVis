```python
import plotly.graph_objects as go
import pandas as pd
import csv
import plotly.express as px
```

```python
df = pd.read_csv('mn17_005aprh25.mat.csv') 
```

```python
startIndex = 32
endIndex = 232
```

```python
xAxis = list(range(startIndex, endIndex))
headData = df.Heading[startIndex:endIndex]
```

```python
head = px.line(x=xAxis, y=headData, labels={'x':'Time', 'y':'Heading'})
head.show()
```

```python
pitchData = df.Pitch[startIndex:endIndex]
```

```python
pitch = px.line(x=xAxis, y=pitchData, labels={'x':'Time', 'y':'Pitch'})
pitch.show()
```

```python
rollData = df.Roll[startIndex:endIndex]
```

```python
roll = px.line(x=xAxis, y=rollData, labels={'x':'Time', 'y':'Roll'})
roll.show()
```

```python
xAccelerationData = df["Accel_X"][startIndex:endIndex]
```

```python
xAccel = px.line(x=xAxis, y=xAccelerationData, labels={'x':'Time', 'y':'Acceleration in the x direction'})
xAccel.show()
```

```python
yAccelerationData = df["Accel_Y"][startIndex:endIndex]
```

```python
yAccel = px.line(x=xAxis, y=yAccelerationData, labels={'x':'Time', 'y':'Acceleration in the y direction'})
yAccel.show()
```

```python
zAccelerationData = df["Accel_Z"][startIndex:endIndex]
```

```python
zAccel = px.line(x=xAxis, y=zAccelerationData, labels={'x':'Time', 'y':'Acceleration in the z direction'})
zAccel.show()
```

```python
pitchRoll = go.Figure()
pitchRoll.add_trace(go.Scatter(x=xAxis, y=pitchData,
                    mode='lines',
                    name='pitch'))
pitchRoll.add_trace(go.Scatter(x=xAxis, y=rollData,
                    mode='lines',
                    name='roll'))
pitchRoll.show()
```

```python
def saveToHTML():
    head.write_html('heading.html', auto_open=True)
    pitch.write_html('pitch.html', auto_open=True)
    roll.write_html('roll.html', auto_open=True)
    xAccel.write_html('xAccel.html', auto_open=True)
    yAccel.write_html('yAccel.html', auto_open=True)
    zAccel.write_html('zAccel.html', auto_open=True)
    pitchRoll.write_html('pitchroll.html', auto_open=True)
```

```python
saveToHTML()
```
